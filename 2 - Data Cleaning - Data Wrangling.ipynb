```python
## Funcion que genera una lista de n numeros aleatorios enteros dentro del intervalo a,b
def randint_list(n,a,b):
    x = [] #declarar un array vacio
    for i in range (n):
        x.append(np.random.randint(a,b)) #añadir al final del array
    return x   
```

```python
randint_list(25,1,50) # le pido 25 numeros aleatorios entre 1 y 50
```

# Seed - semilla de aleatoriedad

```python
np.random.seed(2022)# de esta manera selecciono la semilla de aleatoriedad
for i in range(5):
    print(np.random.random())
```
