# Agregacion de datos por categoria

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male","Female"]
incone = ["Poor","Middle Class", "Rich"]
```

```python
n = 500

#Creo un array vacio
gender_data = []
incone_data = []

#Uso este for para crear 500 valores aleatorios de genero y clase social
for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    incone_data.append(np.random.choice(incone))
```

```python
gender_data[1:20]
```

```python
incone_data[1:10]
```

```python
#Z -> N(0,1) - Distribucion normal media 0 y desviacion tipica 1
#N(m,n) -> n + s * Z - con esta formula se 
#n es la media de las alturas
#s es la dispersion para arriba y para abajo en el caso de la altura de 30 para arriba y 30 para abajo

height = 160+20*np.random.randn(n)
weight = 65+15*np.random.randn(n)
age = 30+12*np.random.randn(n)
income = 18000+3500*np.random.rand(n)
```

```python
# Creo el dataframe a partir de las variables
# que cree que almacenan los datos de las columnas
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status":  incone_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head()
```

La agrupacion de datos se puede llevar a cabo utilizando una funcion llamada groupby, se creara un objeto llamado groupby que tendra dos atributos, el nombre y el grupo.

```python
#Agrupo los datos de la columna genero
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
#Utilizo el for para imprimir el nombre del grupo y el grupo del objeto grouped_gender
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
#Muestro solamente los datos del genero Femenino
grouped_gender.get_group("Female")
```

```python
#Hacer agrupaciones de mas de una columna
double_group = data.groupby(["Gender","Economic Status"]) 
# Se crean 6 categorias 
```

```python
len(double_group) # se puede contar la cantidad de categorias agrupadas
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

# Operaciones sobre datos agrupados

```python
double_group.sum() #devuelve la suma de los datos de cada clase
```

```python
double_group.size() #devuelve la cantidad de elementos en cada una de las clases
```

```python
double_group.describe() #obtener los estadisticos basicos, me trae de todas las columnas
```

```python
grouped_income = double_group["Income"] 
#tomo una columna para aplicarle la operacion
#describe solamente a ese conjunto
#De esta forma me trae solamente los datos de la columna "income"
#o "ingreso" en español con las clases que ya habia filtrado
```

```python
grouped_income.describe()
```

```python
#Realizo un agregado del dataframe y le aplico las funciones que quiera
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": lambda h:np.mean(h)+2,#uso de una funcion lambda
        "Height": np.std
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std]) 
#para aplicar las funciones personalizadas a cada variable
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```
