# Agregacion de datos

```python
import numpy as np
import pandas as pd
np.random.seed(2023)
```

```python
gender = ["Male","Female"] #Genero
income = ["Poor", "Middle Class", "Rich"] #Clase social
```

```python
n = 500

gender_data = [] #Array vacio para el genero
income_data = [] #Array vacio para la clase social

for i in range(0,n):
    gender_data.append(np.random.choice(gender))#generar un valor aleatorio para genero
    income_data.append(np.random.choice(income))#generar un valor aleatorio para la clase social
    
```

```python
gender_data
```

```python
income_data[1:10] #Mostrar solo los 10 primeros
```

```python
height = 160+30* np.random.randn(n) #altura,
weight = 65 +25* np.random.rand(n) # peso
age = 30 + 12 *np.random.randn(n) #edad
income = 18000 + 3500 * np.random.rand(n) #ingreso promedio
```

```python
#Asigno al dataframe los datos a las columnas que creo en esta funcion
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status":income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head() #veo como quedo
```

# Agrupacion de datos por categoria

```python
grouped_gender = data.groupby("Gender") #Creo una variable para agrupar por genero los datos
```

```python
grouped_gender.groups #Veo como queda ese objeto
```

```python
for names, groups in grouped_gender: #Uso un for para iterar y mostrar los datos agrupados por genero, primero mujeres y despues hombres
    print(names)
    print(groups)
```
